```python
## to create a dataframe  out of dictionary
boston_df=pd.DataFrame(data=boston.data,columns=boston.feature_names)
boston_df
```

```python
boston_df['MEDV']=boston.target
boston_df.head()
```

```python
## preparing the input and output columns
x=boston_df[['RM']]
y=boston_df['MEDV']
x
```

```python
y
```

```python
boston_df['MEDV']
```

```python
##plt.style.available
with plt.style.context("grayscale"):
    plt.plot(x,y,'bo')
    plt.xlabel("RM I/P")
    plt.ylabel("MEDV O/P")
    plt.title("scatter plot")
    
```

```python
## usual sklearn : import ,instantiate,fit
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(x, y)
```

```python
## print intercept and coeffecient
print(lm.intercept_)
print(lm.coef_)
```

### Interpret the mode:
A unit increase in x(RM) is associated with a 9.10210898
unit increase in MEDV

```python
##pipeline---> first we create a model on the data
## second we need to compare the model(predict values with actual values)
```

```python
## with help of predict ,we are going to find out predicted values
y_pred=lm.predict(x)
y_pred
```

```python
## compare the actual and predit values
```

```python
## r2_value---> to check the model performnace
from sklearn.metrics import r2_score
r2_score(y,y_pred)
```

```python
## observe the regression line
```

```python
with sns.axes_style("dark"):
    sns.regplot(x,y)
```

```python
## scatter plot regression
plt.scatter(x,y,color='red')
plt.plot(x,y_pred,color='black')
plt.xlabel("RM I/P")
plt.ylabel("MEDV O/P")
```
