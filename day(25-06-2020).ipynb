```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## loading boston dataset
from sklearn.datasets import load_boston
boston=load_boston()
```

```python
boston_df=pd.DataFrame(boston.data,columns=boston.feature_names)
boston_df.head()
```

```python
boston_df['MEDV']=boston.target
boston_df.head()
```

```python
## to find the correlation we use corr()
boston_df.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(boston_df.corr(),annot=True,cmap='YlGnBu')
```

```python
correlation=boston_df.corr()
cor=correlation[correlation>0.7]
plt.figure(figsize=(8,8))
sns.heatmap(cor,annot=True,cmap='YlGnBu')
```

```python
cor=correlation[(correlation>=0.2) & (correlation<=0.5)]
cor
```

```python
## split the data into input  nd output
x=boston_df.drop(['MEDV'],axis=1)
y=boston_df.MEDV
x
```

```python
## divide the x and y into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
## build the model on training data--->x_train,y_train
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(x_train,y_train)
```

```python
## intercept and coeff values
print(lm.intercept_)
print(lm.coef_)
```

```python
## create a dataframe for coefficients
coefficients=pd.DataFrame([x_train.columns,lm.coef_]).T
coefficients
```

```python
## checking the model performance
y_train_pred=lm.predict(x_train)
y_train_pred
```

```python
## we need to compare the actual values(y_train) and the actual values
```

```python
from sklearn.metrics import r2_score
print('R^2:',r2_score(y_train,y_train_pred))
print("Adjusted R^2:",1-(1-r2_score(y_train,y_train_pred))*(len(x_train)-1)/
                                                           (len(x_train)-x_train.shape[1]-1))
```

```python
## regplot (y train-->actual values and y_train_pred---> predicted values)
sns.regplot(y_train,y_train_pred)
```

```python
plt.scatter(y_train,y_train_pred)
plt.xlabel("price")
plt.ylabel("model predicted price")
```

```python
plt.scatter(y_train_pred,y_train-y_train_pred)
plt.xlabel("predicted values")
plt.ylabel("residuals")
```

```python
## test the model on testng data
y_test_pred=lm.predict(x_test)
y_test_pred
```

```python
## we ned to compare the actual values(y_test) and predicted values(y_test_pred)
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print('R^2:',r2_score(y_test,y_test_pred))
print("Adjusted R^2:",1-(1-r2_score(y_test,y_test_pred))*(len(x_test)-1)/
                                                           (len(x_test)-x_test.shape[1]-1))


print("MAE:",mean_absolute_error(y_test,y_test_pred))
print("MSE:",mean_squared_error(y_test,y_test_pred))
print("RMSE:",np.sqrt(mean_squared_error(y_test,y_test_pred)))
```
